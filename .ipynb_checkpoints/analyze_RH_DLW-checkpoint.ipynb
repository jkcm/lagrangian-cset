```python
# RSTB:long_name = "Radiometric Surface Temperature" ;        
# VISB:long_name = "Pyrgeometer Raw Visible Irradiance, Bottom" ;
# IRBC:long_name = "Corrected Infrared Irradiance, Bottom" ;
# IRTC:long_name = "Corrected Infrared Irradiance, Top" ;
# RHODT:long_name = "Absolute Humidity, T-Electric" ;
# RHUM:long_name = "Relative Humidity" ;
# SOLEL:long_name = "Solar Elevation Angle" ;

def make_flight_plot(RF, era_var, axl_var, axr_var):
    fig, axl = plt.subplots(figsize=(14,6))
#     axr = axl.twinx()
    axr=None
    axl.plot(RF.flight_data.Time, RF.flight_data.GGALT, lw=1, c='k')
    axl.set_ylabel("altitude (m)")
    axl.set_ylim(0, 4000)
    X = np.broadcast_to(RF.ERA_data.time.values[:,None], RF.ERA_data[era_var].values.shape)
    pc = axl.pcolormesh(X, RF.ERA_data.ERA_z.values/9.81, RF.ERA_data[era_var].values, cmap='plasma_r', vmin=0, vmax=100)
    sc = axl.scatter(RF.flight_data.Time.values, RF.flight_data.GGALT.values,
                c=RF.flight_data[axl_var].values, cmap='plasma_r', vmin=0, vmax=100)
#     axr.plot(RF.flight_data.Time, RF.flight_data[axr_var])
#     axr.set_ylabel(RF.flight_data[axr_var].long_name + " (" + RF.flight_data[axr_var].units +")")
    plt.colorbar(sc, label=(RF.flight_data[axl_var].long_name + " (" + RF.flight_data[axl_var].units +")"), orientation='horizontal')
    axl.text(0, 0.8, RF.name, horizontalalignment='left', transform=axl.transAxes, fontdict = {'size': 40})
    return fig, axl, axr
                                   
f1, axl1, axr1 = make_flight_plot(RF06, era_var='ERA_r', axl_var='RHUM', axr_var='IRTC')
f2, axl2, axr2 = make_flight_plot(RF07, era_var='ERA_r', axl_var='RHUM', axr_var='IRTC')
f3, axl3, axr3 = make_flight_plot(RF10, era_var='ERA_r', axl_var='RHUM', axr_var='IRTC')
f4, axl4, axr4 = make_flight_plot(RF11, era_var='ERA_r', axl_var='RHUM', axr_var='IRTC')




fs, axls, axrs = [f1], [axl1], [axr1]

fs = [f1, f2, f3, f4]
axls = [axl1, axl2, axl3, axl4]
axrs = [axr1, axr2, axr3, axr4]

myFmt = mdates.DateFormatter('%HZ')
for ax in axls:
    ax.xaxis.set_major_formatter(myFmt)


# for ax in axrs:
#     ax.set_ylim(0, 550)
#     ax.grid(True)
    
def mark_case_edge(flight_piece, ax):
    ax.get_ylim()
    flight_piece.start_time
    ax.axvline(x=flight_piece.start_time, c='r', lw=3)
    ax.axvline(x=flight_piece.end_time, c='r', lw=3)

mark_case_edge(ALC_RF06.outbound_Flight_Piece, axl1)
mark_case_edge(ALC_RF06.return_Flight_Piece, axl2)
mark_case_edge(ALC_RF10.outbound_Flight_Piece, axl3)
mark_case_edge(ALC_RF10.return_Flight_Piece, axl4)

axl1.axvline(x=dt.datetime(2015, 7, 17, 18, 5), c='c', lw=3)
axl1.axvline(x=dt.datetime(2015, 7, 17, 17, 45), c='c', lw=3)
axl1.axvline(x=dt.datetime(2015, 7, 17, 17, 25), c='c', lw=3)
axl1.axvline(x=dt.datetime(2015, 7, 17, 17, 5), c='c', lw=3)

axl2.axvline(x=dt.datetime(2015, 7, 19, 17, 42), c='c', lw=3)
axl2.axvline(x=dt.datetime(2015, 7, 19, 17, 48), c='c', lw=3)
axl2.axvline(x=dt.datetime(2015, 7, 19, 18, 10), c='c', lw=3)
axl2.axvline(x=dt.datetime(2015, 7, 19, 18, 35), c='c', lw=3)

axl3.axvline(x=dt.datetime(2015, 7, 27, 17, 0), c='c', lw=3)
axl3.axvline(x=dt.datetime(2015, 7, 27, 16, 40), c='c', lw=3)
axl3.axvline(x=dt.datetime(2015, 7, 27, 16, 20), c='c', lw=3)

axl4.axvline(x=dt.datetime(2015, 7, 29, 19, 40), c='c', lw=3)
axl4.axvline(x=dt.datetime(2015, 7, 29, 20, 1), c='c', lw=3)
axl4.axvline(x=dt.datetime(2015, 7, 29, 21, 8), c='c', lw=3);

names = ['RF06', 'RF07', 'RF10', 'RF11']
for f, n in zip(fs, names):
    f.tight_layout()
#     f.savefig(os.path.join(utils.plot_dir, '{}_RH_plot.png'.format(n)), dpi=300)
```

```python
#PART 2: check the FT q along all trajectories
import glob, os, utils, 
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_files = glob.glob(r'/home/disk/eos4/jkcm/Data/CSET/Lagrangian_project/trajectory_files/rf*forward*.nc')
qs, lats, lons = [], [], []
for f in all_files:
    data = xr.open_dataset(f)
    qs.append(data.ERA_q.sel(level=700).values*1000)
    lats.append(data.lat.values)
    lons.append(data.lon.values)
qs = np.vstack(qs)
lats = np.vstack(lats)
lons = np.vstack(lons)    
```

```python
#load ERA mean data
all_ERA_files = glob.glob(r'/home/disk/eos4/jkcm/Data/CSET/ERA5/ERA5.pres.NEP.2015-*.nc')
data = xr.open_mfdataset(all_ERA_files)
T = data['t'].sel(level=700).values 
RH = data['r'].sel(level=700).values
p = np.broadcast_to(700, T.shape)*100
q = utils.qv_from_p_T_RH(p, T, RH)*1000
```

```python
fig, [ax1, ax2] = plt.subplots(ncols=2, figsize=(11, 4))
m = utils.bmap(ax=ax1, llr={'lat': (20, 45), 'lon': (-160, -120)},
              par_labs=[1, 0, 0, 0], mer_labs=[0, 0, 1, 0])
xx, yy = np.meshgrid(data.longitude-360, data.latitude)
pc = m.pcolormesh(xx, yy, q.mean(axis=0), latlon=True, vmin=3, vmax=6)
cb = plt.colorbar(pc, ax=ax1, orientation='horizontal')
cb.set_label('$q_{700}$ (g kg$^{-3}$)\n July/Aug 2015 mean')
m.plot(np.mean(lons, axis=0), np.mean(lats, axis=0), c='k', lw=5, latlon=True, zorder=10, label='mean')
m.plot(lons.T, lats.T, alpha=0.5, latlon=True)
ax1.legend()
ax2.plot(qs.T, alpha=0.5)
ax2.plot(np.mean(qs, axis=0), c='k', lw='5', zorder=10);
fig.subplots_adjust(wspace=0.4)
ax2.set_ylabel('$q_{700}$ (g kg$^{-3}$)')
ax2.set_xlabel('time (hours)')
ax2.grid(True)
```

Comparing to Sandu et al., (2010), on which Sandu and Stevens (2011) and Yamaguchi et al. (2017) are based, their trajectory composite starts at around 25N, 125W, and goes to 15N, 138W, so significantly south and east of our trajectories. The resulting FT q gradient experienced would therefore be stronger, and figure 1 in Sandu and Stevens (2011) indicates their their start/end values for q_700 are about 3.5 to 4.5 g/kg over the course of 3 days.
