```python
import datetime as dt
import numpy as np
import os
import xarray as xr
import utils
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = 12,5
import matplotlib.pyplot as plt
import glob
import pandas as pd
from utils import read_tdump
from itertools import cycle
import lagrangian_case as lc
%load_ext autoreload
%autoreload 2
```

```python
ALC_1 = lc.AircraftCase("ALC_RF06A/RF07BCDE")
ALC_2 = lc.AircraftCase("ALC_RF10A/RF11DE")
```

```python
vartoplot = 'CONCU100_RWOOU'
case1_out_var = ALC_1.outbound_flight_data[vartoplot]
case1_ret_var = ALC_1.return_flight_data[vartoplot]

if isinstance(case1_out_var, np.ma.MaskedArray):
    case1_out_var, case1_ret_var = case1_out_var.filled(np.nan), case1_ret_var.filled(np.nan)

bins = np.linspace(0, 200, 11)
#bins = 10

out_hist, bins = np.histogram((case1_out_var)[~np.isnan((case1_out_var))], bins=bins, density=True)
ret_hist, retbins = np.histogram(case1_ret_var[~np.isnan(case1_ret_var)], bins=bins, density=True)
bin_cents = (bins[1:] + bins[:-1])/2

fig, [ax, ax2, ax3] = plt.subplots(nrows=3)
ax.plot(bin_cents, out_hist, label='out')
ax.plot(bin_cents, ret_hist, label='ret')
ax2.plot(case1_out_var, label='out')
ax2.plot(case1_ret_var, label='ret')
ax3.plot(ALC_1.outbound_flight_data['GGALT'], label='out')
ax3.plot(ALC_1.return_flight_data['GGALT'], label='ret')
ax.legend()
ax2.legend()
```

```python
start_times = ALC_1.outbound_legs['Start']
end_times = ALC_1.outbound_legs['End']
legs = ALC_1.outbound_legs['leg']
sequences = ALC_1.outbound_legs['seq']
data=ALC_1.outbound_flight_data

sequence_array = np.empty(data.Time.values.shape, dtype='U1')
leg_array = np.empty(data.Time.values.shape, dtype='U1')
data['leg'] = (('Time'), leg_array)
data['sequence'] = (('Time'), sequence_array)

for s, e, leg, seq in zip(start_times, end_times, legs, sequences):
    which_times = np.logical_and(utils.as_datetime(data['Time'].values) >= s, 
                   utils.as_datetime(data['Time'].values) <= e)
    data['leg'][which_times] = leg
    data['sequence'][which_times] = seq
```
