```python
def special_reshape(arr, mean=False):
    # custom reshape for trajectory data in a silly format. Don't worry about it.
    # input arr is nx1, where n is # of trajectories
    # each entry is a mx1 array, where m is USUALLY 70, but sometimes only ~55 or so (missing data)
    # this inconsistent length is the problem, and why it wasn't 1800x70 in the first place
    # we're gonna deal with this by packing them with nans so they're all 70
    n, m = len(arr), max(arr[0].shape)
    new = np.full((n, m), np.nan)
    for i, t in enumerate(arr):
        if mean:
            t = np.nanmean(t, axis=(0,1))
        new[i,0:len(t.squeeze())] = t.squeeze()
    return new

def get_solar_zenith(lats, lons, hours, dates=None, decs=None):
    if dates:
        #TODO calculate declination here
        pass
    else:
        if not decs:
            raise ValueError("please provide either dates or declination")
    local_hr = hours + 12*lons/180
    hour_angle = 15*(local_hr-12)
    cos_zen = np.sin(np.deg2rad(lats))*np.sin(np.deg2rad(decs)) + \
              np.cos(np.deg2rad(lats))*np.cos(np.deg2rad(decs))*np.cos(np.deg2rad(hour_angle))
    zenith_angle = np.rad2deg(np.arccos(cos_zen))
    return(zenith_angle)
```

```python
def load_climo():
    with open(os.path.join(utils.project_dir, "var_list_onlycf.pickle"), 'rb') as f:
       cf = pickle.load(f)

    with xr.open_dataset(glob.glob(os.path.join(utils.GOES_source, 'G15V03.0.NH*.NC'))[0]) as ds:
       lats = ds.latitude.values
       lons = ds.longitude.values
       
    with open(os.path.join(utils.project_dir, "var_list_nocf_nogamma.pickle"), 'rb') as f:
        data = pickle.load(f)
        
    data.update(cf)
    data.update({"lats": lats, "lons": lons})
    
    mat = loadmat(os.path.join(utils.project_dir, "isabel_data", "Complete_ERA_EIS_LCL_LTS_1000_2015_78_CSET.mat"))
    lats = mat['era_lat']
    lons = mat['era_lon']
    lat_indx = np.logical_and(lats>utils.latlon_range['lat'][0], lats<utils.latlon_range['lat'][1]).flatten()
    lon_indx = np.logical_and(lons>utils.latlon_range['lon'][0], lons<utils.latlon_range['lon'][1]).flatten()
    EIS = mat['EIS_1000'][np.ix_(lon_indx, lat_indx)].mean(axis=2).T
    LTS = mat['LTS_1000'][np.ix_(lon_indx, lat_indx)].mean(axis=2).T
    LCL = mat['LCL_1000'][np.ix_(lon_indx, lat_indx)].mean(axis=2).T
    era_lats = np.broadcast_to(lats[lat_indx], EIS.shape)
    era_lons = np.broadcast_to(lons[lon_indx].T, EIS.shape)
    
    era_dict={"era_lons": era_lons, "era_lats": era_lats, "EIS": EIS, "LTS": LTS, "LCL": LCL}
    data.update(era_dict)    
    
    return data


def convertstring(x):
    return x[0:9] + '{:02d}'.format(int(x[9:11])-1) + x[11:] 



class TrajectoryVariable:
    def __init__(self, traj_arr, clim_arr, good_trajs, hicf_trajs, del_start, del_end):
        
        self.good_raw_traj_arr = traj_arr[good_trajs.astype(bool),:]
        self.hicf_raw_traj_arr = traj_arr[hicf_trajs.astype(bool),:]
        self.good_raw_traj_mean = np.nanmean(self.good_raw_traj_arr, axis=0)
        self.hicf_raw_traj_mean = np.nanmean(self.hicf_raw_traj_arr, axis=0)
        self.good_raw_traj_std = np.nanstd(self.good_raw_traj_arr, axis=0)
        self.good_raw_traj_25pct = np.nanpercentile(self.good_raw_traj_arr, 25, axis=0)
        self.good_raw_traj_75pct = np.nanpercentile(self.good_raw_traj_arr, 75, axis=0)

        
        self.good_clim_traj_arr = clim_arr[good_trajs.astype(bool),:]
        self.hicf_clim_traj_arr = clim_arr[hicf_trajs.astype(bool),:]
        self.good_clim_traj_mean = np.nanmean(self.good_clim_traj_arr, axis=0)
        self.hicf_clim_traj_mean = np.nanmean(self.hicf_clim_traj_arr, axis=0)
        self.good_clim_traj_std = np.nanstd(self.good_clim_traj_arr, axis=0)
        
        self.good_anom_traj_arr = self.good_raw_traj_arr - self.good_clim_traj_arr
        self.hicf_anom_traj_arr = self.hicf_raw_traj_arr - self.hicf_clim_traj_arr
        self.good_anom_traj_mean = np.nanmean(self.good_anom_traj_arr, axis=0)
        self.hicf_anom_traj_mean = np.nanmean(self.hicf_anom_traj_arr, axis=0)
        
        
        self.good_clim_mean = np.nanmean(self.good_clim_traj_arr, axis=1)
        self.hicf_clim_mean = np.nanmean(self.hicf_clim_traj_arr, axis=1)
        self.good_raw_mean = np.nanmean(self.good_raw_traj_arr, axis=1)
        self.hicf_raw_mean = np.nanmean(self.hicf_raw_traj_arr, axis=1)
        self.good_anom_mean = np.nanmean(self.good_anom_traj_arr, axis=1)
        self.hicf_anom_mean = np.nanmean(self.hicf_anom_traj_arr, axis=1)
        
        self.good_raw_delta = self.good_raw_traj_arr[:,del_end] - self.good_raw_traj_arr[:,del_start]
        self.hicf_raw_delta = self.hicf_raw_traj_arr[:,del_end] - self.hicf_raw_traj_arr[:,del_start]
        self.good_clim_delta = self.good_clim_traj_arr[:,del_end] - self.good_clim_traj_arr[:,del_start]
        self.hicf_clim_delta = self.hicf_clim_traj_arr[:,del_end] - self.hicf_clim_traj_arr[:,del_start]
        self.good_anom_delta = self.good_anom_traj_arr[:,del_end] - self.good_anom_traj_arr[:,del_start]
        self.hicf_anom_delta = self.hicf_anom_traj_arr[:,del_end] - self.hicf_anom_traj_arr[:,del_start]
        
        self.good_raw_initval = self.good_raw_traj_arr[:,del_start]
        self.hicf_raw_initval = self.hicf_raw_traj_arr[:,del_start]
        self.good_clim_initval = self.good_clim_traj_arr[:,del_start]
        self.hicf_clim_initval = self.hicf_clim_traj_arr[:,del_start]
        self.good_anom_initval = self.good_anom_traj_arr[:,del_start]
        self.hicf_anom_initval = self.hicf_anom_traj_arr[:,del_start]

        self.good_raw_finalval = self.good_raw_traj_arr[:,del_end]
        self.hicf_raw_finalval = self.hicf_raw_traj_arr[:,del_end]
        self.good_clim_finalval = self.good_clim_traj_arr[:,del_end]
        self.hicf_clim_finalval = self.hicf_clim_traj_arr[:,del_end]
        self.good_anom_finalval = self.good_anom_traj_arr[:,del_end]
        self.hicf_anom_finalval = self.hicf_anom_traj_arr[:,del_end]
        
        # calculate the anomaly change from the expected trend
        #for EG CF: expect that for a high init CF, 
        
#        self.good_raw_initdelcorr = utils.nan_correlate(self.good_raw_initval, self.good_raw_delta)
        idx = np.isfinite(self.good_anom_delta) & np.isfinite(self.good_anom_initval)
        self.good_anom_initdelreg = polyfit(self.good_anom_delta[idx], self.good_anom_initval[idx], deg=1)
        a = np.empty_like(self.good_anom_initval)*np.nan
        a[idx] = polyval(self.good_anom_initdelreg, self.good_anom_initval[idx])
        self.good_anom_delpred = a
        self.good_anom_delres = self.good_anom_delta - self.good_anom_delpred
        
        idx = np.isfinite(self.hicf_anom_delta) & np.isfinite(self.hicf_anom_initval)
        self.hicf_anom_initdelreg = polyfit(self.hicf_anom_initval[idx], self.hicf_anom_delta[idx], deg=1)
        a = np.empty_like(self.hicf_anom_initval)*np.nan
        a[idx] = polyval(self.hicf_anom_initdelreg, self.hicf_anom_initval[idx])
        self.hicf_anom_delpred = a
        self.hicf_anom_delres = self.hicf_anom_delta - self.hicf_anom_delpred
#        self.good_raw_     
    
    
def plot_good_bad_trajectories(mat_dict, ax_g=None, subset=1):
    def plot_single(t, m=None, c=None, i=None, z=random.choice([1,2])):
        m.plot(t.lon.values, t.lat.values, c=c, latlon=True, label=t.tnum[0], zorder=z)
        m.plot(t.lon.values[::6], t.lat.values[::6], '.', c=c, latlon=True, lw=0.5, zorder=z)
        m.plot(t.lon.values[0], t.lat.values[0], '*', c='k', latlon=True, ms=2, zorder=z)
        m.plot(t.lon.values[-1], t.lat.values[-1], 's', c=c, latlon=True, ms=2, zorder=z)
        if i is not None:
            plt.annotate(str(i), xy=(t.lon.values[0]+.5, t.lat.values[0]+.5))

        return m

    # %%
    trajectory_folder = utils.trajectory_dir
    flist = glob.glob(os.path.join(trajectory_folder, 'tdump*'))

    good_traj, bad_traj = 0, 0
    file_list = mat_dict['file_list']
    good_list = mat_dict['good_traj']

    if ax_g is None:
        fig_good, ax_g = plt.subplots(figsize=(7, 5))
    
    llr = {'lat': (10, 45), 'lon': (-160, -115)}
    fig_good, ax_g, m_ax_g = utils.make_map_plot(ax=ax_g, llr=llr, projection='cyl',# lon_0=-135,
                                        par_labs=[1, 0, 0, 0], mer_labs=[0, 0, 1, 0])
#    fig_bad, ax_b, m_ax_b = utils.make_map_plot()
    g_count = 0
    random.shuffle(flist)
    for f in flist:
        tdump = utils.read_tdump(f)
        t_group = tdump.groupby('tnum')
        for key in list(t_group.groups.keys()):

            ncfile = r'{}-{:02}-2deg.nc'.format(os.path.basename(f)[5:13], key)
            indx = np.argwhere(file_list == ncfile).squeeze().item()
            good = good_list[indx]
            traj = t_group.get_group(key)
            color = cm.rainbow(np.random.uniform(0,1))
            if good:
                g_count += 1
                if g_count == subset:
                    good_traj += 1
                    plot_single(t=traj, m=m_ax_g, c='k', z=random.choice([1,2]))
                    g_count = 0
            else:
                bad_traj += 1
#                utils.plot_single(t=traj, m=m_ax_b, c=color)
                
                
    plt.tight_layout(h_pad=0, w_pad=0.5, rect=(0.05, 0.00, 1, 0.96))
    fig_good.patch.set_alpha(0)  
#    ax_g.set_title('CSET good trajectories, {} total'.format(good_traj),
#                   y=1.08)
#    ax_b.set_title('CSET bad trajectories, {} total'.format(bad_traj),
#                   y=1.08)
    return fig_good, m_ax_g
#     fig_good.savefig('/home/disk/p/jkcm/good.png')
#    fig_bad.savefig('/home/disk/p/jkcm/bad.png')
    # %%

    
```

```python
import numpy, scipy.optimize

def fit_sin(tt, yy):
    '''Fit sin to the input time sequence, and return fitting parameters "amp", "omega", "phase", "offset", "freq", "period" and "fitfunc"'''
    tt = numpy.array(tt)
    yy = numpy.array(yy)
    ff = numpy.fft.fftfreq(len(tt), (tt[1]-tt[0]))   # assume uniform spacing
    Fyy = abs(numpy.fft.fft(yy))
    guess_freq = abs(ff[numpy.argmax(Fyy[1:])+1])   # excluding the zero frequency "peak", which is related to offset
    guess_amp = numpy.std(yy) * 2.**0.5
    guess_offset = numpy.mean(yy)
    guess = numpy.array([guess_amp, 2.*numpy.pi*guess_freq, 0., guess_offset])

    def sinfunc(t, A, w, p, c):  return A * numpy.sin(w*t + p) + c
    popt, pcov = scipy.optimize.curve_fit(sinfunc, tt, yy, p0=guess)
    A, w, p, c = popt
    f = w/(2.*numpy.pi)
    fitfunc = lambda t: A * numpy.sin(w*t + p) + c
    return {"amp": A, "omega": w, "phase": p, "offset": c, "freq": f, "period": 1./f, "fitfunc": fitfunc, "maxcov": numpy.max(pcov), "rawres": (guess,popt,pcov)}

from scipy.signal import detrend
cf_dt = detrend(CF_var.good_raw_traj_mean)

# N, amp, omega, phase, offset, noise = 500, 1., 2., .5, 4., 3
# #N, amp, omega, phase, offset, noise = 50, 1., .4, .5, 4., .2
# #N, amp, omega, phase, offset, noise = 200, 1., 20, .5, 4., 1
# tt = numpy.linspace(0, 10, N)
# tt2 = numpy.linspace(0, 10, 10*N)
# yy = amp*numpy.sin(omega*tt + phase) + offset
# yynoise = yy + noise*(numpy.random.random(len(tt))-0.5)

tt = np.arange(len(cf_dt))
yynoise = cth_dt
tt2 = np.arange(len(cf_dt))

res = fit_sin(tt, yynoise)
print( "Amplitude=%(amp)s, Angular freq.=%(omega)s, phase=%(phase)s, offset=%(offset)s, Max. Cov.=%(maxcov)s" % res )

# plt.plot(tt, yy, "-k", label="y", linewidth=2)
plt.plot(tt, yynoise, "ok", label="y with noise")
plt.plot(tt2, res["fitfunc"](tt2), "r-", label="y fit curve", linewidth=2)
plt.legend(loc="best")
# plt.show()
```
