```python
import xarray as xr
import utils
import matplotlib.pyplot as plt
import numpy as np
import os
import datetime as dt
import netCDF4 as nc4
import matplotlib.dates as md
import glob
from matplotlib.colors import LogNorm


import lagrangian_case as lc
from CSET_data_classes import CSET_Flight_Piece
%load_ext autoreload
%autoreload 2
```

```python
files = glob.glob(os.path.join(CSET_Flight_Piece.precip_data_location, '*.cdf'))
for f in files:

# if True:
#     f = files[0]

    data = xr.open_dataset(f)
    date = nc4.num2date(data.time.values.astype('timedelta64[ms]').astype('float')/1000, units=data.time.long_name)
#     date = nc4.num2date(data.time.values, units=data.time.long_name)
    
    data.gv_alt[np.where(data.gv_alt.values < 0)] = np.nan
    timestr = 'seconds since '+ utils.as_datetime(date[0]).strftime("%F %T")
    data['time'] = (('time'), nc4.date2num(date, timestr))
    data['time'] = data['time'].assign_attrs({'long_name': timestr})
    
    fig, ax = plt.subplots(figsize=(14,6))
    ax.set_title(os.path.basename(f)[:4])
    pc = ax.pcolor(date, data.height, data.lwf.values.T*3.6,
                   norm=LogNorm(vmin=1e-4, vmax=1e2))
    
    ax.plot(date, data.gv_alt.values)
    ax.set_ylabel('height (m)')
    cb = plt.colorbar(pc)
    cb.set_label("rain rate (mm/h)")
    # ax.plot(date, data.base.values)
    ax.set_ylim(0, 4000);
    fig.savefig(fname=os.path.join(utils.plot_dir, 'precp_{}'.format(os.path.basename(f)[:4])))
    
```

```python
# plot: histogram of 1) max precips and 2) max precip heights for all data from case 6/10 out/back

# load data
ALC06 = lc.AircraftCase(lc.all_cases[6]['ALC_name'])
ALC06.add_precip_data()
ALC10 = lc.AircraftCase(lc.all_cases[10]['ALC_name'])
ALC10.add_precip_data()
```

```python

def plot_precip_from_case(data, savestr):

    good_index = data['leg'] == 'b'
    no_nan_index = np.any(~np.isnan(data['lwf']), axis=1)
    good_index = np.logical_and(good_index, no_nan_index)
    all_precip = data['lwf'][good_index]
    gv_alt = data['gv_alt'][good_index]
    max_precip_i = np.nanargmax(all_precip, axis=1)
    max_precip = all_precip.max(dim='height', skipna=True)
    max_precip_hgt = [data['height'].values[i] for i in max_precip_i]

    fig, [ax, ax2] = plt.subplots(figsize=(14,6), nrows=2, sharex=True,
                          gridspec_kw={'height_ratios':[5,3]})
    plt.subplots_adjust(hspace=0.1)
    pc = ax.pcolor(range(len(all_precip.time.values)), all_precip.height, all_precip.values.T*3.6, #vmin=0, vmax=100,
                  norm=LogNorm(vmin=1e-4, vmax=1e2))
    ax.set_ylabel('height (m)')
    cb = plt.colorbar(pc, ax=list([ax, ax2]))
    cb.set_label("rain rate (mm/h)")
    ax.set_ylim(0, 2000)
    ax.plot(gv_alt, 'k--', label='GV altitude')
    ax.plot(max_precip_hgt, 'k', label='height of max rain rate')
    ax.grid(True)
    jumps = np.argwhere(np.gradient(all_precip.time.values).astype(float)/1e9 > 600).flatten()
    jumps = [i for i in jumps if i+1 not in jumps]
    for jump in jumps:
        ax.axvline(jump, ls='-', c='w', lw=5)
        ax.axvline(jump, ls='--', c='lightcoral', lw=2, label='data break > 10min')
        ax2.axvline(jump, ls='-', c='w', lw=5)
        ax2.axvline(jump, ls='--', c='lightcoral', lw=2)
    ax.legend()
    
    ax2.plot(max_precip.values*3.6, label=None)
    ax2.set_yscale('log')
    ax2.grid(True)
    ax2.set_ylabel("max rain rate (mm/h)");
    fig.suptitle(savestr + " below-cloud precip retrievals", fontsize=20)

    median_rr = np.nanmedian(max_precip*3.6)
    ax2.axhline(median_rr, label='median colmax rain rate: {:0.3f} mm/hr'.format(median_rr), ls='--', c='r')
    mean_rr = np.nanmean(max_precip*3.6)
    ax2.axhline(mean_rr, label='mean: {:0.3f} mm/hr'.format(mean_rr), ls='-.', c='k')
    ax2.set_yticks([1e-3, 1e-2, 1e-1, 1, 10])
    ax2.set_ylim([1e-4, 12])
    ax2.legend()
    ax2.set_xlim([0, len(all_precip.time.values)])
    tlist = [0] + jumps + [len(all_precip.time.values)]
    ticks = np.concatenate([np.arange(j, tlist[i+1], 100) for i, j in enumerate(tlist[:-1])])
    ax2.set_xticks(ticks)
    ax2.set_xticklabels(['{:%H:%M}'.format(i) for i in utils.as_datetime(all_precip.time.values[ticks])], rotation=45);

    fig.savefig(fname=os.path.join(utils.plot_dir, "precip_max_{}.png".format(savestr)))

for data, savestr in [(ALC06.outbound_Flight_Piece.precip_data, "RF06_outbound"),
                      (ALC06.return_Flight_Piece.precip_data, "RF06_return"),
                      (ALC10.outbound_Flight_Piece.precip_data, "RF10_outbound"),
                      (ALC10.return_Flight_Piece.precip_data, "RF10_return")]:
    plot_precip_from_case(data, savestr)
```

```python
# plot hists

def precip_hist(data, ax, name, col):
    good_index = data['leg'] == 'b'
    no_nan_index = np.any(~np.isnan(data['lwf']), axis=1)
    good_index = np.logical_and(good_index, no_nan_index)
    all_precip = data['lwf'][good_index]
    gv_alt = data['gv_alt'][good_index]
    max_precip_i = np.nanargmax(all_precip, axis=1)
    max_precip = all_precip.max(dim='height', skipna=True)
    max_precip_hgt = [data['height'].values[i] for i in max_precip_i]
    weights = np.ones_like(max_precip.values*3.6)/float(len(max_precip.values*3.6))
    ax.hist(max_precip.values*3.6, weights=weights, bins=np.logspace(-3, 2, 30), label=name, alpha=0.5, color=col, )
    mean = np.nanmean(max_precip*3.6)
    ax.axvline(mean, color=col, label='mean: {:0.2f}'.format(mean))
    
    

fig, [ax1, ax2] = plt.subplots(nrows=2, figsize=(10,10))
for (data, ax, savestr, col) in [(ALC06.outbound_Flight_Piece.precip_data, ax1, "RF06_outbound", 'red'),
                      (ALC06.return_Flight_Piece.precip_data, ax1, "RF06_return", 'b'),
                      (ALC10.outbound_Flight_Piece.precip_data, ax2, "RF10_outbound", 'red'),
                      (ALC10.return_Flight_Piece.precip_data, ax2, "RF10_return", 'b')]:
    precip_hist(data, ax, savestr, col)
for ax in (ax1, ax2):
    ax.legend()
    ax.set_xscale('log')
    ax.set_xlabel("colmax rain rate (mm/h)");
    ax.set_ylabel("frac of obs")
    ax.grid(True)


```
