```python
def plot_case(ALC, TLC, ax=None):
    if not ax:
        fig, ax = plt.subplots(figsize=(8.5,5.5))
    else:
        fig = ax.figure
    llr = {'lat': (20, 45), 'lon': (-160, -120)}

    m = utils.bmap(ax=ax, llr=llr)
    
#     ax.text(0.01, 0.9, '{}/{}'.format(ALC.outbound_flight, ALC.return_flight), 
#             horizontalalignment='left', transform=ax.transAxes, fontdict = {'size': 24})
    
    m.plot(ALC.outbound_flight_data['GGLON'], ALC.outbound_flight_data['GGLAT'],
          latlon=True, lw=6, label='outbound')
    m.plot(ALC.return_flight_data['GGLON'], ALC.return_flight_data['GGLAT'],
          latlon=True, lw=6, label='return')
    
    waypoints = utils.get_waypoint_data(TLC.outbound_flight, waypoint_type='b')
    for k in sorted(TLC.trajectory_files.keys()):
        f = TLC.trajectory_files[k]
        with xr.open_dataset(f) as data:
            end = waypoints.loc[float(f[-6:-3])].ret_time.to_pydatetime()
            idx = utils.as_datetime(data.time.values) < utils.as_datetime(end)
            m.plot(data.lon[idx], data.lat[idx], latlon=True, lw=2, ls='-', label='traj '+f[-6:-3])
#             m.plot(data.lon.values[0], data.lat.values[0], 'x', latlon=True, lw=2, c='k', label=None)
#             m.plot(data.lon.values[idx][-1], data.lat.values[idx][-1], 'x', latlon=True, lw=2, c='k', label=None)
            
    
    ax.legend(loc='upper left', ncol=4, bbox_to_anchor=(-0.1, -0.1), handletextpad=0.4, columnspacing=0.2)
    return fig, ax, m

def add_CF_to_plot(m, TLC, trajnum, time, bbox):
    ax = m.ax
    fig = ax.figure
    pos1 = ax.get_position()
    new_bbox = [pos1.x0 + bbox[0]*pos1.width, pos1.y0 + bbox[1]*pos1.height ,  bbox[2]*pos1.width, bbox[3]*pos1.height] 
    axi  = ax.figure.add_axes(new_bbox)
    axi.imshow(TLC.get_goes_data(trajnum)['visible_count'].isel(time=time), cmap='Greys_r')
    axi.get_xaxis().set_visible(False)
    axi.get_yaxis().set_visible(False)
    
    traj_lat = TLC.traj_data[trajnum].lat[time].values
    traj_lon = TLC.traj_data[trajnum].lon[time].values
    
#     xpt,ypt = ax.transAxes.inverted().transform(ax.transData.transform((traj_lon, traj_lat)))
    xpt, ypt = traj_lon, traj_lat
    bbox = axi.get_position()
    xb, yb = (bbox.x0 + bbox.x1)/2, (bbox.y0 + bbox.y1)/2
    xb, yb = ax.transData.inverted().transform(fig.transFigure.transform((xb, yb)))
    dx, dy = xpt-xb, ypt-yb
    ax.arrow(xb, yb, dx, dy, color='k', width=0.2, length_includes_head=True, zorder=10, head_width=1)
    axi.text(0, 1.05, 't={}hr'.format(time), transform=axi.transAxes, color='red')
    
def add_seq_labels_to_plot(m, ALC, location_array):
    location_array.reverse()
    ax = m.ax
    fig = ax.figure
    pos1 = ax.get_position()
    for fp in [ALC.outbound_Flight_Piece, ALC.return_Flight_Piece]:
        data = fp.flight_data
        for seq in fp.sequences:
            location = location_array.pop()
            new_location = pos1.x0 + location[0]*pos1.width, pos1.y0 + location[1]*pos1.height , 
            lat_s, lon_s = data.where(data.sequence==seq, drop=True).GGLAT.values[0], data.where(data.sequence==seq, drop=True).GGLON.values[0]
            lat_e, lon_e = data.where(data.sequence==seq, drop=True).GGLAT.values[-1], data.where(data.sequence==seq, drop=True).GGLON.values[-1]
            xb, yb = ax.transData.inverted().transform(fig.transFigure.transform((new_location[0], new_location[1])))
            m.plot([lon_e, xb], [lat_e, yb], '-k', zorder=10)
            m.plot([lon_s, xb], [lat_s, yb], '-k', zorder=10)
            ax.plot(xb, yb, '.', c='lightcoral', ms=44, zorder=11)

            ax.annotate(seq, xy=(xb, yb), fontsize=18, transform=ax.transAxes, 
                        horizontalalignment='center', verticalalignment='center', zorder=12)

            
# x.where(x.sequence=='B').GGLAT[0]

    
    
#     traj_lat = TLC.traj_data[trajnum].lat[time].values
#     traj_lon = TLC.traj_data[trajnum].lon[time].values
    
# #     xpt,ypt = ax.transAxes.inverted().transform(ax.transData.transform((traj_lon, traj_lat)))
#     xpt, ypt = traj_lon, traj_lat
#     bbox = axi.get_position()
#     xb, yb = (bbox.x0 + bbox.x1)/2, (bbox.y0 + bbox.y1)/2
#     xb, yb = ax.transData.inverted().transform(fig.transFigure.transform((xb, yb)))
#     dx, dy = xpt-xb, ypt-yb
#     ax.arrow(xb, yb, dx, dy, color='k', width=0.2, length_includes_head=True, zorder=10, head_width=1)
#     axi.text(0, 1.05, 't={}hr'.format(time), transform=axi.transAxes, color='red')
    
    
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=2, figsize=(6.4,10))
w, h = 0.25, 0.25
RF06_bboxes = [[0.7, 0.55, w, h],
               [0.55, 0.15, w, h],
               [0.05, 0.4, w, h]]

RF10_bboxes = [[0.45, 0.55, w, h],
               [0.2, 0.45, w, h],
               [0.05, 0.1, w, h]]

_,_,m1 = plot_case(ALC06, TLC06, ax=ax1)
add_seq_labels_to_plot(m1, ALC06, location_array=
                       [(0.5, 0.8),
                       (0.4, 0.6),
                       (0.3, 0.1)])
_,_,m2 = plot_case(ALC10, TLC10, ax=ax2)
add_seq_labels_to_plot(m2, ALC10, location_array=
                       [(0.8, 0.5),
                       (0.4, 0.2),
                       (0.6, 0.2)])



for ax, ann in zip([ax1, ax2], ['a) case L06', 'b) case L10']):
    ax.annotate(ann, xy=(0.03, 0.90), xycoords='axes fraction', backgroundcolor='white', fontsize=18)



fig.tight_layout()
fig.subplots_adjust(hspace=0.3)
for i, (TLC, trajnum, bboxes, ax) in enumerate(zip([TLC06, TLC10], ['2.3', '6.0'], [RF06_bboxes, RF10_bboxes], [m1, m2])):
    for time, bbox in zip([0, 24, 49], bboxes):
        add_CF_to_plot(ax, TLC, trajnum, time, bbox)
        
fig.savefig(os.path.join(utils.plot_dir, "2_RF06_RF10_map.png"), dpi=300, bbox_inches='tight')
```

```python
def plot_case_small(ALC, TLC, ax=None, label="", par_labs=[0, 0, 0, 0], mer_labs=[0, 0, 0, 0]):
    if not ax:
        fig, ax = plt.subplots(figsize=(8.5,5.5))
    else:
        fig = ax.figure
    llr = {'lat': (20, 45), 'lon': (-160, -120)}

    m = utils.bmap(ax=ax, llr=llr, par_labs=par_labs, mer_labs=mer_labs)
    
    ax.text(0.01, 0.85, label+": "+ "/".join(ALC.name[4:].split('-')), 
            horizontalalignment='left', transform=ax.transAxes, fontdict = {'size': 12})
    
    m.plot(ALC.outbound_flight_data['GGLON'], ALC.outbound_flight_data['GGLAT'],
          latlon=True, lw=6, label='outbound')
    m.plot(ALC.return_flight_data['GGLON'], ALC.return_flight_data['GGLAT'],
          latlon=True, lw=6, label='return')
    
    waypoints = utils.get_waypoint_data(TLC.outbound_flight, waypoint_type='b')
    for k in sorted(TLC.trajectory_files.keys()):
        f = TLC.trajectory_files[k]
        with xr.open_dataset(f) as data:
            end = waypoints.loc[float(f[-6:-3])].ret_time.to_pydatetime()
            idx = utils.as_datetime(data.time.values) < utils.as_datetime(end)
            m.plot(data.lon[idx], data.lat[idx], latlon=True, lw=2, ls='-', label='traj '+f[-6:-3])
            
    
#     ax.legend(loc='upper left', ncol=4, bbox_to_anchor=(-0.1, -0.1), handletextpad=0.4, columnspacing=0.2)
    return fig, ax, m


fig, ax = plt.subplots(figsize=(8, 10), nrows=6, ncols=3, sharex=True, sharey=True)
for i in range(1, 19):
    case = combined_cases[i]
    par_labs=[0, 0, 0, 0]
    mer_labs=[0, 0, 0, 0]
    if i%3 == 1:
        par_labs=[1,0,0,0]
    elif i%3 == 0:
        par_labs=[0,1,0,0]
    if i<4: 
        mer_labs=[0,0,1,0]
    if i>15:
        mer_labs=[0,0,0,1]
    _,_,m1 = plot_case_small(case.ALC, case.TLC, ax=ax.flatten()[i-1], 
                             label=str(i), par_labs=par_labs, mer_labs=mer_labs)
plt.subplots_adjust(hspace=0.1, wspace=0.1)
# for axi in ax[:,0]:
#     axi.set_ylabel([30, 45])
fig.savefig(os.path.join(utils.plot_dir, "SUPP1_maps.png"), dpi=300, bbox_inches='tight')
```

```python
def plot_case_with_profiles(CLC):
    fig, ax, m = plot_case(CLC.ALC, CLC.TLC)
    for FP in [CLC.ALC.outbound_Flight_Piece, CLC.ALC.return_Flight_Piece]:
        profs = FP.get_profiles()
        for seq, prof in profs.items():
            m.plot(prof['zi']['lon'], prof['zi']['lat'], '.k', ms=10,
                   latlon=True, zorder=10)
            m.plot(prof['zi']['lon'], prof['zi']['lat'], 'o', ms=20, mfc='none', mew=5,
                   latlon=True, label = FP.flight_name + seq+ ' prof')
            
    ax.legend(loc='upper left', ncol=4, bbox_to_anchor=(-0.1, -0.1), handletextpad=0.4, columnspacing=0.2)
    return fig, ax, m
# plot_case_with_profiles(combined_cases[6])


for case in all_cases.keys():
    fig, ax, _  = plot_case_with_profiles(combined_cases[case])
    ax.annotate('Case L{}'.format(case), xy=(0.03, 0.90), xycoords='axes fraction', backgroundcolor='white', fontsize=20)
    fig.savefig('/home/disk/p/jkcm/plots/cset_lagrangian/model_case_plots/case_{:2.0f}_with_profiles.jpg'.format(case), dppi=300, bbox_inches='tight')

#     fig.savefig('/home/disk/p/jkcm/plots/cset_lagrangian/model_case_plots/case_{:2.0f}_with_profiles.jpg'.format(case), dppi=300, bbox_inches='tight')
```
