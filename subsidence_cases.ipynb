## Plotting the subsidence along the two CSET Lagrangian case studies

Here we're looking at the lower FT subsidence along CSET trajectories for two cases. RF06A, resampled on RF07BCDE (lots of divergence!), experienced a 'classic' transition in cloudiness, along with significant deepening and a strong LTS decrease (23C-18C). In contrast, RF10A, resampled on RF11DE (note less divergence), while also experiencing a comparable LTS decrease (21C-16C), was in a region of less subsidence, started deeper, and did not have a clear cloudiness transition. RF10 was also a 'polluted' case. Looks like a mix of sulfate and dust...Asian origin? (not BC)

```python
import datetime as dt
import numpy as np
import os
import xarray as xr
import utils
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = 12,5
import matplotlib.pyplot as plt
import glob
import pandas as pd
from utils import read_tdump
from itertools import cycle
%load_ext autoreload
%autoreload 2
```

```python
#TODO: the better way to load these would be to look at which
#      trajecetories start during e.g. "RF10A", but for now...
case_1_trajectories = {'flight': 'rf06', 
                       'trajs': ['3.5', '4.0', '4.3', '4.6', '5.0']}
case_2_trajectories = {'flight': 'rf10', 
                       'trajs': ['5.0', '5.5', '6.0']}

for case in [case_1_trajectories, case_2_trajectories]:
    t_data = []
    for num in case['trajs']:
        tdump = utils.load_flight_trajectory(case['flight'], float(num))
        ds = xr.Dataset.from_dataframe(tdump)
        var_list = ['w', 'r']
        lats, lons, times = ds.lat.values, ds.lon.values, ds.dtime.values
        pressures = np.full_like(lats, fill_value=700)
        ERA_ds = utils.get_ERA_data(var_list=var_list, lats=lats, lons=lons, times=times,
                                   pressures=pressures, box_degrees=2)
        ds = xr.merge([ds, ERA_ds.rename({'w': 'ERA_w', 'r': 'ERA_RH', 'time': 'dtime'})])
        t_data.append(ds)
        print('done with {} {}'.format(case['flight'], num))
    case['data'] = t_data
        
```

```python
fig, ax = plt.subplots()
#casename = case_1_trajectories['trajs'][0]
#case_data = case_1_trajectories['data'][0]
#color = 'r'
#ax=ax
#variable='ERA_w'
def add_traj_to_plot(variable, tdata, plot_dict, ax):
    start_hour = utils.as_datetime(tdata.dtime.values[0]).hour
    offset = start_hour-21
    ax.plot(np.arange(len(tdata.dtime))+offset, tdata[variable], **plot_dict)
    
lines = ["-","--","-.",":"]

for case, col in zip([case_1_trajectories, case_2_trajectories], ['r', 'b']):
    to_mean = []
    linecycler = cycle(lines)
    for (tname, tdata) in zip(case['trajs'], case['data']): 
        if (case['flight'], tname) not in [('rf06', '4.3'), ('rf10', '5.0')]:
            pass
        plot_dict = {"color": col, 
                     "ls": next(linecycler),
                     "label": case['flight'] + "_" + tname,
                     "alpha": 0.5}
        add_traj_to_plot(variable='ERA_w', tdata=tdata, plot_dict=plot_dict, ax=ax)
        to_mean.append(tdata['ERA_w'].values)
    case_mean = np.array(to_mean).mean(axis=0)
    case_std = np.array(to_mean).std(axis=0)
    #ax.fill_between(np.arange(len(case_mean)), case_mean+case_std, case_mean-case_std, color=col, alpha=0.5)
ax.legend()
ax.set_title('red is rf06 (transition), blue is rf10 (deep but stayed cloudy)'+
            '\n offset from local noon on first day')
ax.set_ylabel('w (Pa/s)')
ax.set_xlabel('hours')
ax.grid()
## TODO shift it so that day/night is consistent between them. shift so that 3z is the 0 hour.
```

So it looks like RF06A has consistent 700hpa subsidence of ~0.05 Pa/s, which using hydrostatic eqn is ~0.005m/s, or 5mm/s, which is what we expect. RF10A has highly variable subsidence along the sequence, with the 700hpa vertical wind of the southwestermost trajectory (5.0) actually being upward. General consistent subsidence begins after about 44 hours only.
